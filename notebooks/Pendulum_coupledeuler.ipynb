Pendulum problem using Finite Difference

```python
import math
import numpy
from matplotlib import pyplot as plt
```

```python
#create dictionaries for velocity and angle
v = dict()
theta = dict()
```

```python
#constants

g = 9.8
l = 1
ld = math.sqrt(l/g)
dt = 0.1
```

```python
#exact analytical solution(only for exact angles)

def ang(t):
    ang = (v[0] / ld) * math.sin(ld * t) + theta[0] * math.cos(ld * t)
    return ang
```

```python
#initialize  v and theta

v[0] = 0
theta[0] = 5
```

```python
#difference equations
for i in range(0,999):
    v[i + 1] = v[i] - (ld ** 2) * theta[i] * dt 
    theta[i + 1] = theta[i] + v[i] * dt
```

```python
t = [0.1 * i for i in range(0,1000)]

theta_an = [ang(t[i]) for i in range(0,1000)]
theta_an
```

```python
fig = plt.gcf()
fig.set_size_inches(9,5)
plt.plot(t,theta_an, label = "Analytical")
plt.plot(t, theta.values(),label = "Numerical first-order")
plt.legend()
plt.xlabel('Time-sec')
plt.ylabel('Velocity- m/s')
```
