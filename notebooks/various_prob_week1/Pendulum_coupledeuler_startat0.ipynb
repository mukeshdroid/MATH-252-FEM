# Pendulum problem using Finite Difference

```python
import math
import numpy
from matplotlib import pyplot as plt
```

```python
#create dictionaries for velocity and angle
v = dict()
theta = dict()
```

```python
#constants

g = 9.8
l = 1
ld = math.sqrt(g/l)
dt = 0.01
```

```python
#exact analytical solution(only for exact angles)

def ang(t):
    ang = (v[0] / ld) * math.sin(ld * t) + theta[0] * math.cos(ld * t)
    return ang
```

```python
#initialize  v and theta

v[0] = math.sqrt(2*g*l*(1 - math.cos(math.radians(4))))
theta[0] = 0
```

```python
#difference equations
for i in range(0,999):
    v[i + 1] = v[i] - (ld ** 2) * theta[i] * dt 
    theta[i + 1] = theta[i] + v[i] * dt
```

```python
theta_deg = [(i * 180 / 3.141592) for i in theta.values()]
```

```python
t = [0.01 * i for i in range(0,1000)]

theta_an = [ang(t[i]) for i in range(0,1000)]
theta_an_deg = [(i * 180 / 3.141592) for i in theta_an]
```

```python
plt.plot(t,v.values())
```

```python
fig = plt.gcf()
fig.set_size_inches(9,5)
plt.plot(t,theta_an_deg, label = "Analytical")
plt.plot(t, theta_deg,label = "Numerical first-order")
plt.legend()
plt.xlabel('Time-sec')
plt.ylabel('theta - degrees')
```
