```python
##Given Values

#Cross section areas in m^2
A1 = 0.01
A2 = 0.01
A3 = 0.01

#Youngs Modulus in Pascals

E = 210 * 10 **(9)

#Length in m
L1 = 1
L2 = 1
L3 = math.sqrt(2)

#angle of the elements
T1 = 0
T2 = math.pi/2
T3 = math.pi/4

#forces in N
P = 100000
```

```python
#Generating general stiffness matricesfor a given angle

def stiff(ang):
    c2 = (math.cos(ang))**2
    cs = math.sin(ang) * math.cos(ang)
    s2 = (math.sin(ang))**2
    return np.array([[c2 , cs , -c2 , -cs],
                    [cs , s2 , -cs , -s2],
                    [-c2 , -cs , c2 , cs],
                    [-cs , -s2 , cs , s2]])
```

```python
stiff(0)[0][0]
```

```python
#generating elemnt wise stiffness matrix

K1 = ((E * A1) / L1) * stiff(T1)
K2 = ((E * A2) / L2) * stiff(T2)
K3 = ((E * A3) / L3) * stiff(T3)
```
